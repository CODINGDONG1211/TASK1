Question 1 cleaning the data set

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv')
data.drop_duplicates(inplace=True)
```

REMOVING COLUMN WHICH WAS USELESS AND  ALMOST HAD SAME VALUE

```python
data.drop(['Photo','Flag','Club Logo','Real Face'],axis=1,inplace=True)
```

```python
x=data.iloc[0:,:-1].values
print(x[:,28:50])
```

```python
import numpy as np

# Define a function to evaluate expressions and return the result as an integer
def evaluate_expression(expr):
    try:
        if isinstance(expr, str):
            # Use the eval function to evaluate the expression and convert it to an integer
            result = int(eval(expr))
            return result
        else:
            return None
    except (SyntaxError, ZeroDivisionError, NameError, ValueError):
        # Handle potential errors in the expressions
        return None

# Assuming you have already converted the specified columns to strings
# Create a copy of the specified columns to store the evaluated results
evaluated_values = np.empty(x[:, 28:50].shape, dtype=np.float64)

# Apply the evaluation function to the specified columns
for i in range(evaluated_values.shape[0]):
    for j in range(evaluated_values.shape[1]):
        evaluated_values[i, j] = evaluate_expression(x[i, 28 + j])

# Print the NumPy array with the evaluated results for the specified columns
print(evaluated_values)

x[:,28:50]=evaluated_values
```

Replacing missing values

```python
from sklearn.impute import SimpleImputer

#Here we are using mean method to replace the missing value we can also use most_frequent or median or some standard valued that we would like
imputer =SimpleImputer(missing_values=np.nan,strategy="mean")
x[:,28:50]=imputer.fit_transform(x[:,28:50])
print(x[:,28:50])
```
